###  2ος τρόπος  - Train - test split by scratch ~ returns numpy arrays


```python
from random import seed
from random import randrange
seed(42)

# Split a dataset into a train and test set
def train_test_split(dataset, split=1-split_percent):
    train = list()
    train_size = split * len(dataset) # 0.8 * len of dataset to get train size
    dataset_copy = list(dataset) # copying the list of numpy dataset into the variable dataset_copy
    #print(dataset_copy) -> [array([ 320053, 4394029]), array([ 185853, 7860119],...)] -> [array([ x, y])]
    
    while len(train) < train_size: # while train list is less than train size ..
        index = randrange(len(dataset_copy)) # get an indice from dataset copy which is a list of the dataset
        train.append(dataset_copy.pop(index)) #insert that indice while at the same time you remove it from dataset_copy
    return train, dataset_copy

train, test = train_test_split(df1.to_numpy())
print(len(train))
print(len(test))

```

# Δημιουργία ενός απλού γραμμικού μοντέλου με χρήση  SGD 

### predict function for single linear model

```python
# Make a prediction with coefficients
def predict(row, coefficients):
    yhat = coefficients[0] # yhat = b0
    yhat += coefficients[1] * row[0] # yhat = b0 + b1*x
    return yhat # return yhat value
```

### Εκτίμηση  coefficients με τη χρήση  stochastic gradient descent

Learning Rate: Used to limit the amount each coefficient is corrected each time it is updated.

Epochs: The number of times to run through the training data while updating the coefficients.

3 Loops to implement:
   
   - Loop over each epoch
   - Loop over each row/sample in the training data for an epoch
   - Loop over each coefficient and update it for a row/sample in an epoch
   
Notes:

    error = prediction - expected
    
Update rules: (t is for noting the epoch) -> Remember chain rule theta_L/theta_y_pred -> theta_y_pred/theta_b0=1 kai theta_y_pred/theta_b1 = x1

    b1(t+1) = b1(t) - learning_rate * error(t) * x1(t)
    
    b0(t+1) = b0(t) - learning_rate * error(t)

```python
# Estimate linear regression coefficients using stochastic gradient descent
def coefficients_sgd(train, l_rate, n_epoch,tol):
    coef = [0.0 for i in range(len(train[0]))]
    for epoch in range(n_epoch):
        sum_error = 0
        for row in train:
            yhat = predict(row, coef)
            error = yhat - row[-1]
            #print(error,row[0])
            sum_error += error**2
            coef[0] = coef[0] - l_rate * error
            for i in range(len(row)-1): # den einai kai aparaititi i loupa efoson exw single linear regr.
                coef[i + 1] = coef[i + 1] - l_rate * error * row[i]
        if sum_error <= tol:
            break
        print('>epoch=%d, lrate=%.3f, error=%.3f' % (epoch, l_rate, sum_error))
    return coef
```

```python
# dataset =  list(train_z1.to_numpy()) , from before
l_rate = 0.007
n_epoch = 50
coef = coefficients_sgd(dataset, l_rate, n_epoch,0.001)
print(coef)
```

### MSE για το  test set

```python
Image(url= "https://cdn-media-1.freecodecamp.org/images/hmZydSW9YegiMVPWq2JBpOpai3CejzQpGkNG",width=200,height=200)
```

```python
def mse_metric(actual, predicted):
    sum_error = 0.0
    for i in range(len(actual)):
        prediction_error = predicted[i] - actual[i]
        sum_error += (prediction_error ** 2)
    mean_error = sum_error / float(len(actual))
    return mean_error
```
