__Κλάσεις επικοινωνίας και χαρακτηρισμός αν είναι κλειστές ή ανοιχτές:__

(Α,B) -> κλάση κλειστή

(F,G,H)->κλάση κλειστή

(D,E,C)->κλάση ανοιχτή γιατί επικοινωνεί με την (A,B) και την (F,G,H)


Επαναληπτικές(recurrent) καταστάσεις είναι αυτές που όταν ξεκινάμε από εκεί τότε υπάρχει πιθανότητα να ξανα επιστρέψουμε σε αυτήν την κατάσταση. Αλλιώς αν δεν υπάρχει αυτή η δυνατότητα τότε μιλάμε για παροδικές καταστάσεις(transient).

__Recurrent States:__ F, G, H , A , B

__Transient States:__ E , D, C

Ουσιαστικά οι καταστάσεις που ανήκουν σε κλειστές δομές είναι επαναληπτικές εφόσον σίγουρα θα ξανα επιστρέψουν σε αυτές αν ξεκινάνε απο εκεί. Ενώ αυτές που είναι σε ανοιχτές κλάσεις αν ξεκινήσουν απο εκεί τότε μπορεί να πάνε σε μια κλειστή κλάση και να μη ξανα επιστρέψουν πίσω.
